```python
def append(s, a):
    return s + a

def delete(s):
    if len(s) == 0:
        return ''
    return s[:-1]

all_chars = ['d', 'q']

num_attempts = 0
k = 20
s = 'sfsd'
t = 'sfsdqqqd'

success = False
def possible(s, num_attempts): # s -> t
    if num_attempts == k:
        return
    
    for char in all_chars:
        
        s1 = append(s, char)
        num_attempts += 1
        if s1 == t:
            print(s1, num_attempts)
            if num_attempts == k:
                
                success = True
                print(success)
                return
            else:
                success = False
        else:
            possible(s1,  num_attempts)
        num_attempts -= 1
        
        
    s2 = delete(s)
    num_attempts += 1
    if s2 == t:
        if num_attempts == k:
            success = True
            print(success)
            return
        else:
            success = False
    else:
        return possible(s2,  num_attempts)

possible(s, 0)
print(success)
    
```
