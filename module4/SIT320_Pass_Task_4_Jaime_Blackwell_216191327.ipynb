# Module 4 Overview

A tree is a form of graph. In some applications, trees are the most optimal data structure. Binary Search Trees are an example. They are excellent for search optimisation. Trees are able to achieve O(log(n)) complexity for IDS (insertion, deletion and search) operations, but they must be balanced. There are various algorithms designed to address unbalanced trees, such as AVL Trees (named after Adelson-Velsky and Landis) and RB Trees (Red Black).

Data can be stored in arrays or linked lists.
Arrays take O(n) time to insert or delete an element. To search an array takes O(log(n)) time.
For linked-lists it takes O(1) time to insert and delete, but searching takes O(n) time.
The main motivation behind Binary Search Trees (BSTs) is to reduce the searching complexity of Linked Lists from O(n) to O(log(n))
Although Binary Searcg Trees generally achieve this complexity, sometimes they do not. In some situations, IDS operations can become O(n), the same time complexity as for linked-lists.

**What is BST?**

A BST is implemented using linked lists. Each node has a pointer to three other nodes; the parent node, the left child, and right child nodes. Unless the node is a leaf node (which means it has no child nodes) in which case it only has a parent node.

- Node: An element in the tree.
- Root Node: The first (top) node in the tree. The node with no parent.
- Leaves: The last (bottom) nodes in the tree. The nodes with no children (where both child nodes are NIL)
- Left/Right Children: The immediately consecutive nodes for a given node, there are a maximum of two (left and right).
- Descendant of a Node: A descendant is one that comes after the node (is in some way connected to the node through one of it's child nodes)

**How is a BST built?**

For any given node, the left child node must be less than the parent node. The right node has to be more than it's parent node.

**How are Insert, Delete and Search operations implemented in BST?**
- Search: Insertion and deletion operations depend on the search function. To search, we start from the root node and traverse the tree until we find the search key. Otherwise, we return the element that was found last.

- Delete: To delete a node without creating an imbalance is more complicated. When deleting a node, it must be replaced with another node. Which node should be it's replacement? If the node has no children, or only one child, this is an easy decision. However if it has two child nodes, the deleted node must be replaced by it's immediate successor.

**How long do IDS (insert, delete, search) operations take?**

Search is the important operation that insert and delete both depend on. Searching requires traversing down the tree (deciding to go left or right each time) and the number of decisions we have to make is equal to the height (or depth) of the tree which should be equal to log(n).

**Have some proxy for balance**

Maintaining perfect balance is too hard, so instead we can come up with some proxy for balance. This is the idea that if a tree satisfies some condition, then it's pretty balanced. Eg. height balancing property. balance must be between -1 and 1.

**RB Trees**
RB Trees are a type of binary search tree where the balancing function is done automatically. Based on the constraints applied to the nodes in the tree.

- A red node may only have black nodes as the children.
- The root node has to be black as well
- The black nodes has to be balanced.
- The insertion of new nodes have to be red.
- for any node X; for all paths from X to nils have the same number of black nodes on them.

The key takeaway is to think about why they are balanced. They are balanced because of the constraints that ensure the number of black nodes for each path are exactly the same. The red-black structure is a proxy for balance.
