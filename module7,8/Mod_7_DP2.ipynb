### <font color='red'> Activity 1: You are running up a staircase with a total of n steps. You can hop either 1 step, 2 steps or 3 steps at at time. Write a DP program to determine how many possible ways you can run up the stairs? (Hint: Start with a recursive solution, and then later move to top-down approach of DP). </font>



```python
### TODO ###
### Good Luck ###
def countWaysRecursive(n):
    # Base cases
    if n == 1:
        return 1  # One way: 1 hop
    if n == 2:
        return 2  # Two ways: 1-1 hop or 2 hop
    if n == 3:
        return 4  # Four ways: 1-1-1 hop, 1-2 hop, 2-1 hop, or 3 hop
    if n < 1:
        return 0  # No ways for negative steps
    
    # Recursive cases:
    # From step n-1: If you were on step n-1, you could take a single hop to reach step n.
    # The number of ways to get to step n-1 is ways(n-1).
    # From step n-2: If you were on step n-2, you could take a double hop to reach step n. 
    # The number of ways to get to step n-2 is ways(n-2).
    # From step n-3: If you were on step n-3, you could take a triple hop to reach step n. 
    # The number of ways to get to step n-3 is ways(n-3).
    return countWaysRecursive(n-1) + countWaysRecursive(n-2) + countWaysRecursive(n-3)


def countWaysTopDown(n, memo={}):
    if n == 0:
        return 1
    if n < 0:
        return 0
    if n in memo:
        return memo[n]
    memo[n] = countWaysTopDown(n-1, memo) + countWaysTopDown(n-2, memo) + countWaysTopDown(n-3, memo)
    return memo[n]


def countWaysBottomUp(n):
    # If n is less than 1, there are no ways to climb
    if n < 1:
        return 0
    
    # Initialize an array dp of size n+1 to store the number of ways for each step
    dp = [0] * (n+1)
    
    # Base cases
    dp[1] = 1  # One way: 1 hop
    if n >= 2:
        dp[2] = 2  # Two ways: 1-1 hop or 2 hop
    if n >= 3:
        dp[3] = 4  # Four ways: 1-1-1 hop, 1-2 hop, 2-1 hop, or 3 hop
    
    # Fill up the dp array using the bottom-up approach
    for i in range(4, n+1):
        dp[i] = dp[i-1] + dp[i-2] + dp[i-3]
    
    return dp[n]




```

### <font color='red'> Activity 2: Write the code for finding the Longest Common Sub-sequence. Make sure you output the Matrix C and teh longest sub-sequence as well. Test your code with various use-cases. </font>
* Define the subproblem
* What is the initial subproblem?

The Longest Common Subsequence (LCS) problem can be broken down into smaller subproblems using dynamic programming. The subproblem for the LCS can be defined as follows:

Given two sequences `X` and `Y`, let's denote:
- `X[i]` as the prefix of `X` of length `i`.
- `Y[j]` as the prefix of `Y` of length `j`.

The subproblem is to determine the length of the LCS of the prefixes `X[i]` and `Y[j]`.

We can denote this subproblem solution as `L[i][j]`, where `L[i][j]` represents the length of the LCS of `X[i]` and `Y[j]`.

The recursive formulation for the subproblem is:

1. If `i == 0` or `j == 0`, then `L[i][j] = 0` (i.e., one of the sequences is empty).
2. If `X[i-1] == Y[j-1]`, then `L[i][j] = L[i-1][j-1] + 1` (i.e., the current characters match, so the LCS length increases by 1).
3. If `X[i-1] != Y[j-1]`, then `L[i][j] = max(L[i-1][j], L[i][j-1])` (i.e., the current characters don't match, so we take the maximum LCS length from either excluding the current character of `X` or excluding the current character of `Y`).

By solving these subproblems iteratively for all `i` and `j`, we can build up the solution for the original problem, which is the LCS of the entire sequences `X` and `Y`. The solution will be `L[m][n]`, where `m` is the length of `X` and `n` is the length of `Y`.

```python
### TODO ###
### Good Luck ###
import numpy as np

def LCS(X, Y):
    # Initialize a 2D array C to store lengths of common subsequences
    C = np.zeros((len(X)+1, len(Y)+1), dtype=int)
    
    # Initialize a 2D array common_sequences to store the actual subsequences
    common_sequences = np.empty((len(X)+1, len(Y)+1), dtype=object)
    common_sequences.fill('')
    
    # Loop through each character of X and Y
    for i in range(1, len(X)+1):
        for j in range(1, len(Y)+1):
            # print(f"Before: ({i}, {j})", common_sequences[i, j])
            # If the characters match, increment the LCS length and add the character to the subsequence
            if X[i-1] == Y[j-1]:
                C[i, j] = C[i-1, j-1] + 1
                common_sequences[i, j] = common_sequences[i-1, j-1] + X[i-1]
                # print(f"After: ({i}, {j})", common_sequences[i, j])
            else:
                # If the characters don't match, choose the longer subsequence from the previous row or column
                if C[i, j-1] > C[i-1, j]:
                    C[i, j] = C[i, j-1]
                    common_sequences[i, j] = common_sequences[i, j-1]
                else:
                    C[i, j] = C[i-1, j]
                    common_sequences[i, j] = common_sequences[i-1, j]
                # print(f"After: ({i}, {j})", common_sequences[i, j])
                
    
    # Return the longest common subsequence, which is in the bottom-right cell of the common_sequences array
    return common_sequences

# Test 1
result = LCS('ABCDEF', 'BEF')
print(result)
# if result == 'BEF':
#     print('Test 1: Passed')
# else:
#     print('Test 1: Failed')

# # Test 2
# result = LCS('ACATWITHABATANDANICEHAT', 'AFLYWITHAKNIFEANDABALOON')
# if result == 'AWITHAANDAN':
#     print('Test 2: Passed')
# else:
#     print('Test 2: Failed')

```

```python
def lcs_memorization(str1, str2, i, j, memo):
    if i == 0 or j == 0:
        return 0, ""
    
    if (i, j) in memo: # IF THEY are in the dictionary
        return memo[(i, j)]
    
    if str1[i - 1] == str2[j - 1]:
        length, sequence = lcs_memorization(str1, str2, i - 1, j - 1, memo)
        sequence = sequence + str1[i - 1]
        length += 1 #Adding 1 to length seperately
        memo[(i, j)] = (length, sequence)
    else:
        length1, sequence1 = lcs_memorization(str1, str2, i - 1, j, memo)
        length2, sequence2 = lcs_memorization(str1, str2, i, j - 1, memo) 
        if length1 > length2: # kind of doing max here
            memo[(i, j)] = (length1, sequence1)
        else:
            memo[(i, j)] = (length2, sequence2)
    
    return memo[(i, j)]

def lcs(str1, str2):
    m,n= len(str1), len(str2)
    
    
    mem = {} #Dict for memorization
    
    length, seq = lcs_memorization(str1, str2, m, n, mem)
 
    print("Longest Common Subsequence Length:", length)
    
    #memoization matrix
    print("Memoization Matrix:")
    for i in range(m + 1):
        for j in range(n + 1):
            if (i, j) in mem:
                print(mem[(i, j)][0], end=" ")
            else:
                print("None", end=" ")
        print()

    # Printing the LCS sequence
    print("Longest Common Subsequence:", seq)


result = lcs('ABCDEF', 'BEF')

print(result)
```

```python
#### TODO ####
### Good Luck ###



def unboundedKnapsack(W, n, wt, vals, names):
 
    K = [0 for i in range(W + 1)]
    ITEMS = [[] for i in range(W + 1)]
 
    for x in range(1, W + 1):
        K[x] = 0
        for i in range(1, n):
            
            prev_k = K[x]
            
            if (wt[i] <= x):
                K[x] = max(K[x], K[x - wt[i]] + vals[i])
                
            if K[x] != prev_k:
                ITEMS[x] = ITEMS[x - wt[i]] + names[i]
                
 
    return K[W], ITEMS[W]


W = 4
wt = [1, 2, 3]
vals = [1, 4, 6]
names = [["Turtle"], ["Globe"], ["WaterMelon"]]
n = len(vals)
print('We have {} items'.format(n))                
K, ITEMS = unboundedKnapsack(W, n, wt, vals, names)


```

```python
def knapsack_tabular(W, n, wt, vals, names):
    K = [[0] * (W + 1) for i in range(n + 1)]  # [[0 for _ in range(W + 1)] for _ in range(n + 1)]
    ITEMS = [[] for i in range(W + 1)] # REUSING from Unbound Knapsack
    
    for i in range(1, n + 1):
        for w in range(1, W + 1):
            if wt[i - 1] <= w:
                # take the maximum value between skipping the item and including it
                K[i][w] = max(vals[i - 1] + K[i - 1][w - wt[i - 1]], K[i - 1][w])
                if K[i][w] != K[i - 1][w]:
                    ITEMS[w] = ITEMS[w - wt[i - 1]] + [names[i - 1]]
            else:
                # If the current item's weight is greater than the current weight, skip it
                K[i][w] = K[i - 1][w]
    
    for t in K: # Printing a row
        print(t)
    
    return K[n][W], ITEMS[W]



W = 4
wt = [1, 2, 3]
vals = [1, 4, 6]
names = ["Turtle", "Globe", "WaterMelon"]

n = len(vals)
print('We have {} items'.format(n))                



max_value, selected_items = knapsack_tabular(W, n, wt, vals, names)
print("Selected items are:", selected_items)
print("The maximum value is :", max_value)
```
